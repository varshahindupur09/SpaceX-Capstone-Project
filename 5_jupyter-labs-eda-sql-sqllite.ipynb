### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
%sql select distinct "Customer" from SPACEXTBL; 
```

```python
%sql select sum("PAYLOAD_MASS__KG_") as total_payload_mass_nasa_crs from SPACEXTBL where "Customer" = "NASA (CRS)"; 
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
%sql select distinct "Booster_Version" from SPACEXTBL; 
```

```python
%sql select avg("PAYLOAD_MASS__KG_") as avg_payload_mass_bv_f9v1_1 from SPACEXTBL where "Booster_Version" = "F9 v1.1"; 
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
%sql select distinct "Landing_Outcome" from SPACEXTBL; 
```

```python
%sql select "Date" from SPACEXTBL where "Landing_Outcome" = "Success (ground pad)" order by "Date" limit 1;
```

```python
%sql select min("Date") from SPACEXTBL where "Landing_Outcome" = "Success (ground pad)";
```

P.S. The use of ORDER BY ensures that the entire dataset is arranged in ascending order based on the "Date" column. This guarantees that the date shared is not only the first in the list but is also in the correct chronological order.

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
%sql select "Booster_Version" from SPACEXTBL where "PAYLOAD_MASS__KG_" between 4000 and 6000 and "Landing_Outcome" = "Success (drone ship)" 
```

```python
# df.columns.tolist() # ['Date','Time (UTC)','Booster_Version','Launch_Site','Payload', 'PAYLOAD_MASS__KG_',
# 'Orbit','Customer','Mission_Outcome','Landing_Outcome']
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
%sql select "Landing_Outcome", count(*) as landing_count from SPACEXTBL where "Date" between "2010-06-04" and "2017-03-20" group by "Landing_Outcome" order by landing_count DESC;
```

```python
%sql SELECT "Landing_Outcome", COUNT(*) AS landing_count, RANK() OVER (ORDER BY COUNT(*) DESC) AS ranking from SPACEXTBL WHERE "Date" BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY "Landing_Outcome" ORDER BY ranking; 
```
